## Kmeans with Random Sampling

K means algorithem with ordering based on overal class occurences in whole image and random sampling

```python
def find_n_best_colors_sorted(n, image_array):
    image_array = np.array(image_array, dtype=np.float64) / 255 # convert to float and scale 0 - 1
    w, h, d = original_shape = tuple(image_array.shape) # width, height, color depth
    image_array = np.reshape(image_array, (w * h, d)) # reshape into list of individual
    
    image_array_sample = shuffle(image_array, random_state=0)[:1000] # sample 1000 random pixels
    kmeans = KMeans(n_clusters=n, random_state=0).fit(image_array_sample) # find n clusters
    centroids = kmeans.cluster_centers_

    ## Sort by most common
    unique_l, counts_l = np.unique(kmeans.labels_, return_counts=True)
    sort_ix = np.argsort(counts_l)
    sort_ix = sort_ix[::-1]
    
    sorted_centroids = centroids[sort_ix]
        
    return (centroids, kmeans)
```

using html color dictionary see `./helper.py`

```python


def test_img(path,evaluator,args,crop=(0.25,0.25)):
    color_map = get_html_color_dict()
    
    img = Image.open(path)
    
    print('original size', img.size)
    plot_img(img)
    
    img = crop_img(img, crop)
    
    print('cropped size', img.size)
    plot_img(img)
    
    colors, model = evaluator(args, img)
    
    plot_swatches(colors)
    
    named_colors = [get_closest_color(c, color_map) for c in colors]
    
    print(named_colors)
    
    return (colors, named_colors, model)
    
    
```

Alot of clusters are required before the pink is detected, goal will be trying to reduce this

```python
def rand_sampling(width=1, height=1, n=500):
    sample = [(np.random.uniform(0, width), np.random.uniform(0, height)) for _ in range(0, n) ]
    
    return sample
```

```python
def rand_sample_img_array(array, plot=False):
    w, h, d = tuple(array.shape)
    
    print(w,h,d)
    
    sample = rand_sampling(width=w, height=h, n=500)
    sample = np.vectorize(lambda x: int(round(x)))(sample)
    
    sample_pixels = [array[x,y] for (y,x) in sample if x >= 0 and x <= w - 1 and y >= 0 and y <= h - 1]
    
    if plot:
        df_sample = pd.DataFrame(sample, columns=['y', 'x'])
        plot = sns.jointplot(x="x", y="y", data=df_sample)
        
        
    return (sample_pixels, df_sample)
```

```python
img = Image.open('images/flamingo-shirt.jpg')

img = crop_img(img, crop=(0.25,0.25))

img_array = np.array(img)

pixels, samples = rand_sample_img_array(img_array, plot=True)

print('pixels: ', len(pixels))

fig, ax = plt.subplots(figsize=(10,10))

ax.imshow(img)
ax.scatter(samples.x, samples.y)
plt.show()
```

## Dart Sampling

based on implementation: https://www.labri.fr/perso/nrougier/from-python-to-numpy/code/DART_sampling_numpy.py

```python
from scipy.spatial.distance import cdist


def DART_sampling_numpy(width=1.0, height=1.0, radius=0.025, k=100):

    # Theoretical limit
    n = int((width+radius)*(height+radius) / (2*(radius/2)*(radius/2)*np.sqrt(3))) + 1
    # 5 times the theoretical limit
    n = 5*n

    # Compute n random points
    P = np.zeros((n, 2))
    P[:, 0] = np.random.uniform(0, width, n)
    P[:, 1] = np.random.uniform(0, height, n)

    # Computes respective distances at once
    D = cdist(P, P)

    # Cancel null distances on the diagonal
    D[range(n), range(n)] = 1e10

    points, indices = [P[0]], [0]
    i = 1
    last_success = 0
    while i < n and i - last_success < k:
        if D[i, indices].min() > radius:
            indices.append(i)
            points.append(P[i])
            last_success = i
        i += 1
    return points
```

```python
dart_sample = pd.DataFrame(DART_sampling_numpy(), columns=['x', 'y'])
sns.jointplot(x="x", y="y", data=dart_sample)
```

create image sampling method, take 1000 blue noise samples of the image array

```python
def dart_sample_img_array(array, plot=False):
    w, h, d = tuple(array.shape)
    
    print(w,h,d)
    
    sample = DART_sampling_numpy(width=w, height=h, radius=5, k=100)
    sample = np.vectorize(lambda x: int(round(x)))(sample)
    
    sample_pixels = [array[x,y] for (y,x) in sample if x >= 0 and x <= w - 1 and y >= 0 and y <= h - 1]
    df_sample = pd.DataFrame(sample, columns=['y', 'x'])
    if plot:
        plot = sns.jointplot(x="x", y="y", data=df_sample)
        
        
    return (sample_pixels, df_sample)
    
    
```
